```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('betas_table.csv')
df = df.loc[df['date']>'2019-01-01']
df = df.loc[df['symbol']=='JPM US']
df['cumulative_alpha'] = (1 + df['one_day_alpha']).cumprod()
df
```

```python
import pandas as pd
import plotly.express as px

fig = px.line(df, x='date', y='cumulative_alpha',title = 'JPM Cumulative Alpha')
fig.show()
```

```python
fig = px.line(df, x='date', y='spx_return_coef',title = 'JPM 2 yr Daily Rolling Beta')
fig.show()
```

```python
df.loc[df['symbol']=='JPM US']
coef_cols = list(df.columns)
coef_cols = [x for x in coef_cols if 'return_coef' in x]
df_cols = df[coef_cols]
df_cols = pd.DataFrame(df_cols.mean()).reset_index()
df_cols.columns=['factor','value']
df_cols = df_cols.loc[~df_cols['factor'].isin(['earnings_close_return_coef','BAMLH0A0HYM2_return_coef','SP500_return_coef'])]
df_cols.sort_values('value')
fig = px.bar(df_cols, x='factor', y='value',title="JPM Factor Loadings")
fig.update_layout(barmode='stack', xaxis={'categoryorder':'total ascending'})
fig.show()
```

```python
df_plot = df.loc[df['earnings_trade_date']==1]
mean = df_plot['one_day_alpha'].mean()
fig = px.bar(df_plot, x='date', y='one_day_alpha',title="JPM Earnings Day Alpha History")
fig.add_hline(y=mean,line_color="Red")
fig.show()
```
